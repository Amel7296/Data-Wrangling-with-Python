```python
import sqlite3
import pandas as pd
```

```python
df1 = pd.read_csv("persons.csv")
df1.head()
```

```python
conn = sqlite3.connect("petsdb")
```

```python
df1['city'] = df1['city'].map(lambda x: x.strip())
```

```python
df1.to_sql("persons", conn, index=False)
```

```python
df2 = pd.read_sql_query('SELECT * FROM persons', conn)
df2.head()
```
